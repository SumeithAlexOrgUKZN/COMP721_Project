Now the we are going to create the new Dataset

A lambda function is used to enter wheather a player is considered an outstanding player if they have played in an allstar game.

```python
def getClass(x):
    if x in playerAllstarDf.unique():
        return 1
    else: 
        return 0
```

```python
playerCareerDf['class'] = playerCareerDf['ilkid'].apply(lambda x: getClass(x))
```
