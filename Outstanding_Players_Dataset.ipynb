# Current Dataset Breakdown

We created a new dataset to determine the outstanding players.

This was done by ultising the dataset in player_regular_season_career.txt and adding a additional column called class to state wheather they are an outstanding player.

This additional column is deteremined wheater a player played in an allstar match from player_allstar.txt. THis is because players who play in allstar games are regarded as NBA star players.

The datasets are read

```python
playerCareerDf = pd.read_csv('databasebasketball/player_regular_season_career.txt')
playerAllstarDf = pd.read_csv('databasebasketball/player_allstar.txt')
```

We only need the players who played in the allstar games, so we droped the rest of the data.

```python
playerAllstarDf = playerAllstarDf['ilkid']
```

```python
playerAllstarDf = playerAllstarDf.drop_duplicates()
```

```python
playerAllstarDf = playerAllstarDf.apply(lambda x: x.upper().strip()) 
```

```python
playerAllstarDf
```

A lambda function is used to enter wheather a player is considered an outstanding player if they have played in an allstar game.

```python
def getClass(x):
    if x in playerAllstarDf.unique():
        return 1
    else: 
        return 0
```

```python
playerCareerDf['class'] = playerCareerDf['ilkid'].apply(lambda x: getClass(x))
```

The new dataset is saved in a arff file so we can create modules using arff 

```python
playerCareerDf.drop(['firstname', 'lastname', "leag"], axis = 1, inplace = True) 
```

```python
arff.dump('playersData.arff'
      , playerCareerDf.values
      , relation='players'
      , names=playerCareerDf.columns)
```
